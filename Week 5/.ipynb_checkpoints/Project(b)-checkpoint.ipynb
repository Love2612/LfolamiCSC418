```python
import cv2
import tkinter as tk
from tkinter import filedialog, ttk
from PIL import Image, ImageTk
import numpy as np
import os

class PAUObjectDetector:
    def __init__(self, root):
        self.root = root
        self.root.title("PAU Object Detector")
        self.root.geometry("1000x700")

        # YOLO setup
        self.net = cv2.dnn.readNet("cfg/yolov3.weights", "cfg/yolov3.cfg")
        self.classes = []
        with open("cfg/coco.names", "r") as f:
            self.classes = [line.strip() for line in f.readlines()]
        self.layer_names = self.net.getLayerNames()
        self.output_layers = [self.layer_names[i - 1] for i in self.net.getUnconnectedOutLayers()]

        # GUI Setup
        self.setup_gui()

        # Video handling
        self.cap = None
        self.video_paths = []
        self.current_video = 0
        self.is_playing = False
        self.total_frames = 0

    def setup_gui(self):
        bg_color = "#2c3e50"
        button_color = "#3498db"
        text_color = "#ecf0f1"

        self.root.configure(bg=bg_color)

        main_frame = tk.Frame(self.root, bg=bg_color)
        main_frame.pack(fill=tk.BOTH, expand=True, padx=20, pady=20)

        self.video_label = tk.Label(main_frame, bg="black")
        self.video_label.pack(fill=tk.BOTH, expand=True)

        control_frame = tk.Frame(main_frame, bg=bg_color)
        control_frame.pack(fill=tk.X, pady=10)

        btn_style = {'bg': button_color, 'fg': text_color, 'bd': 0, 'padx': 15, 'pady': 8}

        self.load_btn = tk.Button(control_frame, text="Load Videos", command=self.load_videos, **btn_style)
        self.load_btn.pack(side=tk.LEFT, padx=5)

        self.play_btn = tk.Button(control_frame, text="Play", command=self.toggle_play, **btn_style)
        self.play_btn.pack(side=tk.LEFT, padx=5)

        self.prev_btn = tk.Button(control_frame, text="Previous Video", command=self.prev_video, **btn_style)
        self.prev_btn.pack(side=tk.LEFT, padx=5)

        self.next_btn = tk.Button(control_frame, text="Next Video", command=self.next_video, **btn_style)
        self.next_btn.pack(side=tk.LEFT, padx=5)

        self.info_label = tk.Label(control_frame, text="Ready to load videos", bg=bg_color, fg=text_color)
        self.info_label.pack(side=tk.LEFT, padx=20)

        self.video_listbox = tk.Listbox(control_frame, height=3, bg=bg_color, fg=text_color, selectbackground=button_color)
        self.video_listbox.pack(side=tk.RIGHT, padx=5)
        self.video_listbox.bind("<<ListboxSelect>>", self.select_video)

        self.seek_var = tk.DoubleVar()
        self.seek_bar = ttk.Scale(main_frame, from_=0, to=100, orient="horizontal", variable=self.seek_var, command=self.seek_video)
        self.seek_bar.pack(fill=tk.X, padx=10, pady=5)

        self.progress = ttk.Progressbar(main_frame, mode='determinate', length=300)
        self.progress.pack(pady=5)

        self.count_label = tk.Label(main_frame, text="Objects Detected: 0", bg=bg_color, fg=text_color, font=("Arial", 12))
        self.count_label.pack()

    def load_videos(self):
        self.video_paths = []
        video_dir = "videos"
        if os.path.exists(video_dir):
            for file in os.listdir(video_dir):
                if file.endswith((".mp4", ".avi", ".mov")):
                    self.video_paths.append(os.path.join(video_dir, file))

        self.video_listbox.delete(0, tk.END)
        for path in self.video_paths:
            self.video_listbox.insert(tk.END, os.path.basename(path))

        self.current_video = 0
        if self.video_paths:
            self.info_label.config(text=f"Loaded {len(self.video_paths)} videos")
        else:
            self.info_label.config(text="No videos found in 'videos' folder")

    def select_video(self, event):
        selection = event.widget.curselection()
        if selection:
            self.current_video = selection[0]
            self.play_video()

    def next_video(self):
        if not self.video_paths:
            return
        self.current_video = (self.current_video + 1) % len(self.video_paths)
        self.play_video()

    def prev_video(self):
        if not self.video_paths:
            return
        self.current_video = (self.current_video - 1) % len(self.video_paths)
        self.play_video()

    def toggle_play(self):
        self.is_playing = not self.is_playing
        self.play_btn.config(text="Pause" if self.is_playing else "Play")
        if self.is_playing:
            self.update_frame()

    def play_video(self):
        if not self.video_paths:
            return
        if self.cap:
            self.cap.release()
        self.cap = cv2.VideoCapture(self.video_paths[self.current_video])
        self.total_frames = int(self.cap.get(cv2.CAP_PROP_FRAME_COUNT))
        self.progress["maximum"] = self.total_frames
        self.is_playing = True
        self.update_frame()

    def update_frame(self):
        if not self.is_playing or not self.cap or not self.cap.isOpened():
            return

        ret, frame = self.cap.read()
        if not ret:
            self.cap.set(cv2.CAP_PROP_POS_FRAMES, 0)
            return

        frame = self.detect_objects(frame)
        frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        img = Image.fromarray(frame_rgb)
        img = ImageTk.PhotoImage(image=img)

        self.video_label.img = img
        self.video_label.config(image=img)

        current_frame = int(self.cap.get(cv2.CAP_PROP_POS_FRAMES))
        self.seek_var.set((current_frame / self.total_frames) * 100)
        self.progress["value"] = current_frame

        self.root.after(10, self.update_frame)  # Reduced delay for smoother playback

    def seek_video(self, value):
        if self.cap and self.cap.isOpened():
            frame_number = int((float(value) / 100) * self.total_frames)
            self.cap.set(cv2.CAP_PROP_POS_FRAMES, frame_number)

    def detect_objects(self, frame):
        height, width, _ = frame.shape
        blob = cv2.dnn.blobFromImage(frame, 0.00392, (416, 416), (0, 0, 0), True, crop=False)
        self.net.setInput(blob)
        outs = self.net.forward(self.output_layers)

        class_ids = []
        confidences = []
        boxes = []
        for out in outs:
            for detection in out:
                scores = detection[5:]
                class_id = np.argmax(scores)
                confidence = scores[class_id]
                if confidence > 0.5:
                    center_x = int(detection[0] * width)
                    center_y = int(detection[1] * height)
                    w = int(detection[2] * width)
                    h = int(detection[3] * height)
                    x = int(center_x - w / 2)
                    y = int(center_y - h / 2)
                    boxes.append([x, y, w, h])
                    confidences.append(float(confidence))
                    class_ids.append(class_id)

        indexes = cv2.dnn.NMSBoxes(boxes, confidences, 0.5, 0.4)
        detected_count = len(indexes)
        self.count_label.config(text=f"Objects Detected: {detected_count}")

        font = cv2.FONT_HERSHEY_PLAIN
        colors = np.random.uniform(0, 255, size=(len(self.classes), 3))
        for i in range(len(boxes)):
            if i in indexes:
                x, y, w, h = boxes[i]
                label = str(self.classes[class_ids[i]])
                confidence = confidences[i]
                color = colors[class_ids[i]]
                cv2.rectangle(frame, (x, y), (x + w, y + h), color, 2)
                cv2.putText(frame, f"{label} {confidence:.2f}", (x, y + 30), font, 2, color, 2)

        return frame

if __name__ == "__main__":
    root = tk.Tk()
    app = PAUObjectDetector(root)
    root.mainloop()
```
