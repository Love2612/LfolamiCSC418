```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread('imgs/cc_and_tt.jpeg')

# Plot the originam image
plt.subplot(1, 2, 1)
plt.title("Original")
plt.imshow(img)

# Scale the image by a factor of 2 along both axes
scaled_img = cv2.resize(img, None, fx=2, fy=2)

# Save the image
cv2.imwrite('imgs/Scaled.jpeg', scaled_img)

# Plot the scaled image
plt.subplot(1, 2, 2)
plt.title("Scaled")
plt.imshow(scaled_img)
plt.show()
```
