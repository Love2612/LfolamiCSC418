```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

img = cv2.imread('imgs/c_and_t.jpeg')

# Plot the originam image
plt.subplot(1, 2, 1)
plt.title("Original")
plt.imshow(img)

# Inverse by subtracting from 255
inverse_img = 255 - img

# Save the image
cv2.imwrite('imgs/Inverse-image.jpeg', inverse_img)

# Plot the scaled image
plt.subplot(1, 2, 2)
plt.title("Inverse colour")
plt.imshow(inverse_img)
plt.show()
```
